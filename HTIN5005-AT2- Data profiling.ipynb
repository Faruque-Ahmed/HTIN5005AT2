# Load  Patient Admission dataset

```python
import pandas as pd
##### Load  Patient Admission dataset
data = pd.read_csv('./admission.csv')
data
```

```python
df=data
cols=df.columns
print("List of columns")
for col in cols:
    print(col)
```

# Generate the Basic Data Profile Report

```python
# Get basic information
print("Basic Information:")
print(df.info())

# Get descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe(include='all'))

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

# Get the distribution of disposition(Target) variable

```python

# Get the distribution of disposition(Target) variable
print(df['disposition'].value_counts())
```

# Heatmap on Demographics and Triage   

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select the columns of Demographics and Triage Variables for the heatmap   
columns = [
    "age", "gender", "ethnicity", "race", "lang", "religion", "maritalstatus",
    "employstatus", "insurance_status", "dep_name", "arrivalmode", "arrivalmonth",
    "arrivalday", "arrivalhour_bin", "triage_vital_hr", "triage_vital_sbp",
    "triage_vital_dbp", "triage_vital_rr", "triage_vital_o2",
    "triage_vital_o2_device", "triage_vital_temp", "esi"
]

# Create a correlation matrix
corr_matrix = df[columns].apply(lambda x: pd.factorize(x)[0]).corr()

# Plot the heatmap
plt.figure(figsize=(15,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
